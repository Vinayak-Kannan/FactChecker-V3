```python
import pandas as pd

df = pd.read_parquet("hf://datasets/tdiggelm/climate_fever/data/test-00000-of-00001.parquet")
```

```python
df
```

```python
# Filter out claims where df is not 0 or 1 in claim_label
df = df[df["claim_label"].isin([0, 1])]

ground_truth = pd.DataFrame(columns=["text", "Numerical Rating"])
rows_to_add = []

for index, row in df.iterrows():
    if row["claim_label"] == 0:
        rows_to_add.append({"text": row["claim"], "Numerical Rating": 3})
    else:
        rows_to_add.append({"text": row["claim"], "Numerical Rating": 1})

# Convert rows_to_add to a DataFrame and concatenate it with ground_truth
ground_truth = pd.concat([ground_truth, pd.DataFrame(rows_to_add)], ignore_index=True)

# Loop through evidences value in the df dataframe. Convert the evidences value to a list of dictionaries
# where each dictionary contains the text and the label of the evidence
evidences = []
for index, row in df.iterrows():
    for evidence in row["evidences"]:
        evidences.append({"text": evidence["evidence"], "Numerical Rating": 3})
print(evidences)

# Convert evidences to a DataFrame and concatenate it with ground_truth
ground_truth = pd.concat([ground_truth, pd.DataFrame(evidences)], ignore_index=True)
```

```python
# Print value counts for numerical rating in ground_truth
print(ground_truth["Numerical Rating"].value_counts())
```

```python
ground_truth.to_csv("ground_truth.csv", index=False)
```
