```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

tvec = TfidfVectorizer(stop_words='english')
dataframe = pd.read_csv('Transformed Data/clustered_claims.csv')

documents = dataframe['claim'].tolist()
cluster = dataframe['cluster'].tolist()
# Get unique clusters
cluster = np.unique(cluster)
merged_docs = []
for i, c_num in enumerate(cluster):
    cluster_docs = dataframe[dataframe['cluster'] == c_num]['claim'].tolist()
    merged_docs.append(' '.join(cluster_docs))

tvec_weights = tvec.fit_transform(merged_docs)
print(tvec_weights.shape)
```

```python
def top_tfidf_feats(row, features, top_n=25):
    ''' Get top n tfidf values in row and return them with their corresponding feature names.'''
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids]
    df = pd.DataFrame(top_feats)
    df.columns = ['feature', 'tfidf']
    return df

output = top_tfidf_feats
```
